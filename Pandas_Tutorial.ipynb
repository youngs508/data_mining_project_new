```python
import pandas as pd
energy_consumption = pd.read_csv('C:\\Users\\Public\\housing_price.csv', index_col=0)
```

```python
energy_consumption.drop(['id', 'date', 'view', 'condition', 'grade', 'zipcode', 'lat', 'long'],
             axis=1,           
             inplace=True)
energy_consumption
```

```python
ec = energy_consumption.loc[:, ['bedrooms', 'bathrooms', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'price']]
ec
```

```python
ec
```

```python
ec.loc[:, 'year of construction'] = 2024 - ec.loc[:, 'yr_built']
for i in range(len(ec)):  
    if ec.loc[i, 'yr_renovated'] != 0:
        ec.loc[i, 'year of construction'] = 2024 - ec.loc[i, 'yr_renovated'] 
        
ec
```

```python
ec.loc[:, 'livable space (sqft)'] = ec.loc[:, 'sqft_above']
ec.loc[:, 'basement (sqft)'] = ec.loc[:, 'sqft_basement']

# energy consumption updated
ec_new = ec.loc[:, ['bedrooms', 'bathrooms', 'livable space (sqft)', 'basement (sqft)', 'year of construction', 'price']]
ec_new
```

```python
ec_new2 = ec_new.sort_values(by=['price'])
ec_new2
```

```python
ec_new3 = ec_new2.loc[ec_new2['year of construction'] <= 40]
ec_new3
```
